### Step 3: Chunk the Document

```python
def find_natural_boundary(chunk, prefix_len, min_size):
    """Find the best natural boundary cut point within a chunk.
    
    Returns the cut index (relative to chunk start) or None if no
    suitable boundary is found.
    """
    # 1. Try paragraph boundary (double newline)
    idx = chunk.rfind("\n\n")
    if idx != -1 and prefix_len + idx + 2 >= min_size:
        return idx + 2  # include the double newline

    # 2. Fall back to sentence boundary (". ")
    idx = chunk.rfind(". ")
    if idx != -1 and prefix_len + idx + 2 >= min_size:
        return idx + 2  # include the period and space

    # 3. Fall back to word boundary (single space — cut after the last complete word)
    idx = chunk.rfind(" ")
    if idx != -1 and prefix_len + idx + 1 >= min_size:
        return idx + 1  # include the space; next chunk starts at the following word

    return None

def compute_distance_from_text_end(start, prefix_len, text_len):
    """Compute the distance from the current start position to the end of the text."""
    return text_len - start - prefix_len

def compute_end(start, max_size, prefix_len, text_len):
    """Compute the candidate end index for the current chunk window."""
    distance_to_text_end = compute_distance_from_text_end(start, prefix_len, text_len) # Calculate the remaining chunk size from start to end of text (adjusted for prefix)
    idx_end_small_chunk = start + distance_to_text_end - prefix_len  # Calculate the end index if we were to take the entire remaining text as a chunk (adjusted for prefix)
    
    return min(start + max_size - prefix_len, idx_end_small_chunk)

def apply_boundary(start, end, text, prefix, min_size):
    """Attempt to snap the end of the chunk to a natural boundary.
    
    Returns the adjusted end index, capped at len(text) if end overshoots.
    Never cuts mid-word: if the cut point falls inside a word, advances to
    the next whitespace boundary (or end of text).
    """
    text_len = len(text)
    text_end_reached = False

    if end >= text_len:
        text_end_reached = True
        return text_len, text_end_reached  # Cap to end of text — no boundary snapping needed

    chunk = text[start:end]
    cut = find_natural_boundary(chunk, len(prefix), min_size)

    if cut is not None:
        end = start + cut

    # Advance end forward until it lands on a whitespace boundary or end of text,
    # so we never split mid-word.
    while end < text_len and not text[end].isspace():
        end += 1

    return end, text_end_reached


def build_chunk(prefix, text, start, end):
    """Concatenate the carried-forward prefix with the current text slice."""
    return prefix + text[start:end + 1] # Include the character at end index since advance_start will move past it


def handle_small_chunk(new_chunk, chunks, end, text_len, min_size, max_size):
    """Handle a chunk that is below the minimum size threshold.
    
    Returns (should_return, pending):
      - should_return: True if the caller should immediately return chunks
      - pending: the chunk to carry forward, or None
    """
    if end >= text_len:
        # At end of text — merge into previous chunk or emit as-is
        if chunks and len(chunks[-1]) + len(new_chunk) <= max_size:
            chunks[-1] += new_chunk
        else:
            chunks.append(new_chunk)
        return True, None

    # Too small mid-text — carry forward as prefix for the next chunk
    return False, new_chunk


def snap_to_word_start(position, text):
    """Advance position forward until it points to the start of a word.

    If position falls mid-word (i.e. the character before it is non-whitespace),
    advance forward until we reach a whitespace character, then skip past it so
    the returned position is the first character of the next word.
    If position is already at a word boundary (preceded by whitespace or at
    position 0), it is returned unchanged.
    """
    text_len = len(text)

    if position == 0 or position >= text_len:
        return position

    # If we're mid-word, advance to the end of the current word
    if not text[position - 1].isspace():
        while position < text_len and not text[position].isspace():
            position += 1

    # Skip past the whitespace so we land on the first character of the next word
    while position < text_len and text[position].isspace():
        position += 1

    return position


def advance_start(start, end, overlap, text):
    """Compute the next start position, always beginning at a word boundary.

    Advances to end - overlap, but never backward from current start.
    If that produces no meaningful forward progress (i.e. the overlap
    region has already been passed), jump directly to end instead.
    The resulting position is then snapped forward to the nearest word start
    so chunks never begin mid-word.
    """
    next_start = end - overlap
    if next_start <= start:
        # overlap region already behind us — skip past end entirely
        next_start = end

    return snap_to_word_start(next_start, text)


def chunk_text(text, max_size=500, overlap=50, min_size=250):
    """Split text into overlapping chunks with natural boundaries.

    Tries to end chunks at paragraph boundaries first, then sentence boundaries,
    then word boundaries, then falls back to max_size. Only breaks at a boundary
    if the resulting chunk is >= min_size characters.

    Small chunks (< min_size) are merged with the next chunk to maximize size,
    unless the previous chunk can absorb them without exceeding max_size.

    Chunks always begin at a word boundary — never mid-word.
    """
    chunks = []
    start = 0
    pending = None  # small chunk carried forward

    while start < len(text):
        prefix = pending if pending is not None else ""
        pending = None

        end = compute_end(start, max_size, len(prefix), len(text))
        end, text_end_reached = apply_boundary(start, end, text, prefix, min_size)

        new_chunk = build_chunk(prefix, text, start, end)

        if text_end_reached:
            # At end of text — emit chunk and return
            chunks.append(new_chunk)
            return chunks

        elif len(new_chunk) < min_size:
            should_return, pending = handle_small_chunk(
                new_chunk, chunks, end, len(text), min_size, max_size
            )
            if should_return:
                return chunks
        else:
            chunks.append(new_chunk)

        start = advance_start(start, end, overlap, text)

    return chunks
```

### Step 5: Perform Dimensionality Reduction to 2D for Visualization and Create Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import umap
```

```python

"""
# Perform dimensionality reduction to 2D for visualization
"""
# Reduces high-dimensional embeddings to 2D using UMAP, with a n_jobs=1 
# for reproducibility
reduced_embeddings = umap.UMAP(n_components=2, n_jobs=1).fit_transform(np.array(embeddings))


"""
# --- Identify two clusters via KMeans ---
"""
# Imports the KMeans clustering algorithm from scikit-learn
from sklearn.cluster import KMeans

# Initializes KMeans to find 2 clusters, with a fixed seed and 10 different \
# centroid initializations for stability

    # KMeans is a clustering model, which is an unsupervised learning algorithm \
    # that finds groupings in data without any pre-labeled examples. It discovers \
    # structure on its own, with no "correct answer" to learn from. 
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)

# Fits KMeans to the reduced embeddings and returns a cluster label (0 or 1) \
# for each point
labels = kmeans.fit_predict(reduced_embeddings)

# Defines two colors (red and blue) to visually distinguish the two clusters
colors = ['#e74c3c', '#3498db']

# Maps each point's cluster label to its corresponding color
cluster_colors = [colors[label] for label in labels]

# Creates a new matplotlib figure with an 8x6 inch canvas
plt.figure(figsize=(8, 6))

# Plots each reduced embedding as a scatter point, colored by cluster, with \
# slight transparency and size 50
plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1],
            c=cluster_colors, alpha=0.7, s=50)

# Iterates over each point's index and (x, y) coordinates to annotate each point \
# with its index number
for i, (x, y) in enumerate(reduced_embeddings):
    # Adds the index label slightly offset (5, 5 points) from each scatter point in 
    # small, semi-transparent text
    plt.annotate(str(i), (x, y), textcoords="offset points", xytext=(5, 5), fontsize=7, alpha=0.8)



"""
# Legend
"""
# Imports the Patch class to create colored rectangle legend entries
from matplotlib.patches import Patch

# Creates two legend patches, one per cluster, each with its assigned color and label
legend_elements = [Patch(facecolor=colors[0], label='Cluster 0'),
                   Patch(facecolor=colors[1], label='Cluster 1')]

# Adds the legend to the plot, letting matplotlib choose the best position
plt.legend(handles=legend_elements, loc='best')

# Sets the chart title describing the plot content
plt.title("UMAP Projection of Chunk Embeddings")

# Adjusts subplot parameters so all elements fit within the figure without overlap
plt.tight_layout()

# Renders and displays the completed plot
plt.show()




"""
# Group chunks by cluster
"""
# Collects all text chunks whose KMeans label is 0 into a list
cluster_0_chunks = [chunks[i] for i, label in enumerate(labels) if label == 0]

# Collects all text chunks whose KMeans label is 1 into a list
cluster_1_chunks = [chunks[i] for i, label in enumerate(labels) if label == 1]




"""
# Defines a function that sends a cluster's chunks to GPT-4o-mini and prints a \
short thematic description
"""
def describe_cluster(cluster_chunks, cluster_id):
    # Joins up to the first 10 chunks with double newlines to form a single \
    # prompt-ready string, avoiding token limit overruns
    combined_text = "\n\n".join(cluster_chunks[:10])  # limit to 10 chunks to stay within token limits

    # Sends the combined chunk text to GPT-4o-mini, asking it to summarize the \
    # unifying theme in 2-3 sentences
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            # Sets the assistant's role as a text analysis and summarization helper
            {"role": "system", "content": "You are a helpful assistant that analyzes and summarizes groups of text chunks."},
            # Provides the combined chunks and instructs the model to identify the 
            # common topic or theme
            {"role": "user", "content": f"Here are text chunks that were grouped together by semantic similarity:\n\n{combined_text}\n\nIn 2-3 sentences, describe what topic or theme unites these chunks."}
        ]
    )
    # Extracts the model's text response from the API result
    description = response.choices[0].message.content

    return cluster_id, description

# Calls describe_cluster for cluster 0, passing in its chunks and ID
chunks_0, description_0 = describe_cluster(cluster_0_chunks, 0)
# Calls describe_cluster for cluster 1, passing in its chunks and ID
chunks_1, description_1 = describe_cluster(cluster_1_chunks, 1)


# Print the descriptions for each cluster with proper formatting and safety checks
import textwrap
from IPython import get_ipython
import shutil

# Get terminal width safely (works in notebooks, consoles, etc.)
try:
    width = shutil.get_terminal_size().columns
except OSError:
    width = 88  # Fallback if terminal size can't be determined

# Wrap and print descriptions (with safety checks)
def safe_fill(text, width):
    if isinstance(text, str):
        return textwrap.fill(text, width=width)
    else:
        return f"Warning: Description is not a string (got {type(text).__name__}). Check describe_cluster output."

print(f"Cluster 0 description:\n{safe_fill(description_0, width)}\n")
print(f"Cluster 1 description:\n{safe_fill(description_1, width)}\n")
```
